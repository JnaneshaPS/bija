```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import files
uploaded = files.upload()
df = pd.read_csv("concrete_data.csv")
df.head()
```

```
print("Average Cement content:", df['Cement (kg/m3)'].mean())
print("Maximum Strength:", df['Strength (MPa)'].max())
print("Minimum Water content:", df['Water (kg/m3)'].min())
```

```
df['w/c'] = df['Water (kg/m3)'] / df['Cement (kg/m3)']
df
```

```
plt.figure(figsize=(6,4))
plt.bar(df['Mix_ID'], df['Strength (MPa)'], color='skyblue')
plt.xlabel('Mix ID')
plt.ylabel('Strength (MPa)')
plt.title('Concrete Strength for Different Mixes')
plt.show()
```

```
plt.figure(figsize=(6,4))
sns.scatterplot(x='w/c', y='Strength (MPa)', data=df, s=100, color='red')
plt.title('Effect of w/c Ratio on Strength')
plt.show()
```
